```python
import pandas as pd
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("Mechanical_datasets/1985_Automobile_dataset/auto_clean.csv")
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
bodyStyle_dummies=pd.get_dummies(df['body-style'], drop_first=True)
pd.get_dummies(df['body-style'], drop_first=True)
cylindersNum_dummies=pd.get_dummies(df['num-of-cylinders'], drop_first=True)
```

```python
cylindersNum_dummies = cylindersNum_dummies.rename(columns={'twelve':'twelveCylinder', 'six':'sixCylinder', 'five':'fiveCylinder', 'four':'fourCylinder', 'three':'threeCylinder', 'two':'twoCylinder'})
cylindersNum_dummies.head()
```

```python
driveWheels_dummies = pd.get_dummies(df['drive-wheels'], drop_first=True)
driveWheels_dummies.head()
```

```python
new_df = df.drop(['body-style','num-of-cylinders','drive-wheels'], axis=1)
```

```python
new_df = pd.concat([new_df, bodyStyle_dummies], axis=1)
new_df.head()
```

```python
new_df = pd.concat([new_df, cylindersNum_dummies], axis=1)
new_df.head()
```

```python
new_df = pd.concat([new_df, driveWheels_dummies], axis=1)
new_df.head()
```

```python
df_cpy = new_df
```

```python
input_df = df[['compression-ratio','horsepower']]
input_df.corr()
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler 
scaler = MinMaxScaler()
scaler.fit(input_df[['compression-ratio']])
scaler.fit(input_df[['horsepower']])
input_df['compression-ratio'] = scaler.transform(input_df[['compression-ratio']])
input_df['horsepower'] = scaler.transform(input_df[['horsepower']])
input_df.corr()
```

```python
##KMeans Clustering
km = KMeans(n_clusters=2)
y_predicted = km.fit_predict(input_df)
input_df['cluster']=y_predicted
y_predicted
```

```python
df1 = input_df[input_df.cluster==0]
df2 = input_df[input_df.cluster==1]
plt.scatter(df1['compression-ratio'],df1['engine-size'],color='green')
plt.scatter(df2['compression-ratio'],df2['engine-size'],color='red')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
##Elbow Plot
from matplotlib import pyplot as plt
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(input_df)
    sse.append(km.inertia_)

plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
